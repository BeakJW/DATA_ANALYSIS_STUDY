```python
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
# ν•κΈ€μ„ μ§€μ›ν•λ” ν°νΈ νμΌ
# ν°νΈμ κΈ€κΌ΄μ„ μ‹μ¤ν…μ— λ“±
font_prop = fm.FontProperties(fname=font_path) # ν°νΈμ μ†μ„±μ„ μ½μ–΄μ΄
font_name = font_prop.get_name() # μ½μ–΄μ¨ μ†μ„±μ—μ„ ν°νΈμ μ΄λ¦„λ§ μ¶”μ¶
font_name
```
