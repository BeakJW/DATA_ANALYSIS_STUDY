### ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò

```python
from hossam import load_data
# Í∏ÄÍº¥ÏùÑ ÏãúÏä§ÌÖúÏóê Îì±Î°ù
from matplotlib import font_manager as fm
# Ï∫îÎ≤ÑÏä§(figure)Î•º ÏÉùÏÑ±, Í∏∞Î≥∏ Í∑∏ÎûòÌîΩ Ìï®Ïàò Ï†úÍ≥µ
from matplotlib import pyplot as plt
# Í≥†Í∏â Í∑∏ÎûòÌîΩ Í∏∞Îä• Ï†úÍ≥µ
import seaborn as sb
```

#### ÏãúÏä§ÌÖú Ï†ÑÏó≠ ÏÑ§Ï†ï(Ïª¥Ìì®ÌÑ∞ÎßàÎã§ ÌïúÎ≤à)

```python
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
# ÌïúÍ∏ÄÏùÑ ÏßÄÏõêÌïòÎäî Ìè∞Ìä∏ ÌååÏùº
# Ìè∞Ìä∏Ïùò Í∏ÄÍº¥ÏùÑ ÏãúÏä§ÌÖúÏóê Îì±
font_prop = fm.FontProperties(fname=font_path) # Ìè∞Ìä∏Ïùò ÏÜçÏÑ±ÏùÑ ÏùΩÏñ¥Ïò¥
font_name = font_prop.get_name() # ÏùΩÏñ¥Ïò® ÏÜçÏÑ±ÏóêÏÑú Ìè∞Ìä∏Ïùò Ïù¥Î¶ÑÎßå Ï∂îÏ∂ú
font_name
```

#### Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï(ÌååÏùºÎãπ ÌïúÎ≤à ÏÑ†Ïñ∏)

```python
my_dpi = 120
my_font_name = "Noto Sans KR" 
my_theme = "dark" # Ïù¥ÎØ∏ÏßÄ ÏÑ†Î™ÖÎèÑÎ•º Í≤∞Ï†ïÌïòÎäî 1Ïù∏Ïπò
# ÏãúÏä§ÌÖúÏóê Îì±Î°ùÎêú Í∏ÄÍº¥ Ïù¥Î¶Ñ
# Í∑∏Î¶º Ïä§ÌÉÄÏùº ÏßÄÏ†ï
sb.set_theme(style=my_theme) # seaborn Ïä§ÌÉÄÏùº (ÌôîÌíç ÏÑ§Ï†ïÌïòÍ∏∞)
plt.rcParams['font.family'] = my_font_name # Í∑∏ÎûòÌîÑÏóê ÌïúÍ∏Ä Ìè∞Ìä∏ Ï†ÅÏö©
plt.rcParams['font.size'] = 16 # Í∏∞Î≥∏ Ìè∞Ìä∏ ÌÅ¨Í∏∞
plt.rcParams['axes.unicode_minus'] = False # Í∑∏ÎûòÌîÑÏóê ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ(ÌïúÍ∏ÄÌôòÍ≤ΩÏóêÏÑú ÌïÑÏàò)üìò
```

### üìò #02. Í∑∏ÎûòÌîÑ Í∏∞Î≥∏ ÏΩîÎìú Íµ¨ÏÑ±

```python
# 1) Í∑∏ÎûòÌîÑ Ï¥àÍ∏∞Ìôî (Ï∫îÎ∞îÏä§(fig)ÏôÄ ÎèÑÌôîÏßÄ(ax) Ï§ÄÎπÑÌïòÍ∏∞)
width_px = 1200 # Í∑∏ÎûòÌîÑ Í∞ÄÎ°ú ÌÅ¨Í∏∞
height_px = 760 # Í∑∏ÎûòÌîÑ ÏÑ∏Î°ú ÌÅ¨Í∏∞
rows = 1 # ÎèÑÌôîÏßÄÏùò Ìñâ Ïàò
cols = 1 # ÎèÑÌôîÏßÄÏùò Ïó¥ Ïàò
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)
# Í∑∏ÎûòÌîÑÏùò ÎèÑÌôîÏßÄ ÏÉÅÌÉú ÌôïÏù∏Ïö© ÌÖåÏä§Ìä∏ ÏΩîÎìú
print(ax)
# 2) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ -> seaborn ÏÇ¨Ïö©
# ...
# 3) Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞ -> ÎèÑÌôîÏßÄ(ax)Ïóê ÏßÅÏ†ë Ï†ÅÏö©
# color: Í∏ÄÏûê ÏÉâÏÉÅ (Í∏∞Î≥∏Í∞íÏùÄ black)
# fontsize: Í∏ÄÏûê ÌÅ¨Í∏∞ (Í∏∞Î≥∏Í∞íÏùÄ plt.rcParams['font.size'] ÏÑ§Ï†ïÏùÑ Îî∞Î¶Ñ)
# pad, labelpad : Í∑∏ÎûòÌîÑÏôÄÏùò Í∞ÑÍ≤©
# fontweight: Í∏ÄÏûê ÍµµÍ∏∞ (100~1000 ÏÇ¨Ïù¥ 100Îã®ÏúÑ Í∞í). Í∏ÄÍº¥Ïù¥ ÏßÄÏõêÌïòÎäî Í≤ΩÏö∞Îßå Ï†ÅÏö©Îê®
ax.set_title("Ï†úÎ™©ÏûÖÎãàÎã§", color="#000000", fontsize=22,
fontweight=1000, pad=20)
ax.set_xlabel("XÏ∂ï", color="#000000", fontsize=16, labelpad=10)
ax.set_ylabel("YÏ∂ï", color="#000000", fontsize=16, labelpad=10)
# 4) Ï∂úÎ†•
plt.grid() # Î∞∞Í≤Ω Í≤©Ïûê ÌëúÏãú/Ïà®ÍπÄ (ÌÖåÎßàÏóê Îî∞Îùº Îã§Î¶Ñ)
plt.tight_layout() # Ïó¨Î∞± Ï†úÍ±∞
plt.savefig("myplot.png", dpi=my_dpi) # ÏÉùÎûµ Í∞ÄÎä•
plt.show() # Í∑∏ÎûòÌîÑ ÌôîÎ©¥ Ï∂úÎ†•
plt.close()
```

### #03. Line Plotüìò

#### Í∏∞Î≥∏Í∑∏Î¶¨Í∏∞

```python
# 1) Í∑∏ÎûòÌîÑ Ï¥àÍ∏∞Ìôî (Ï∫îÎ∞îÏä§(fig)ÏôÄ ÎèÑÌôîÏßÄ(ax) Ï§ÄÎπÑÌïòÍ∏∞)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ -> seaborn ÏÇ¨Ïö©
sb.lineplot([10,20,30,40,50])

#3) Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
ax.set_title("Í∏∞Î≥∏Í∑∏Î¶¨Í∏∞", color="#000000", fontsize=22,
fontweight=1000, pad=20)
ax.set_xlabel("XÏ∂ï", color="#000000", fontsize=16, labelpad=10)
ax.set_ylabel("YÏ∂ï", color="#000000", fontsize=16, labelpad=10)

# 4) Ï∂úÎ†•
#plt.grid()
plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # Í∑∏ÎûòÌîÑ ÌôîÎ©¥ Ï∂úÎ†•
plt.close()
```

#### üìù 2. x, yÏ∂ï Î∞è ÏÑ† Î™®Ïñë ÏßÄÏ†ïÌïòÍ∏∞

```python
# 1) Í∑∏ÎûòÌîÑ Ï¥àÍ∏∞Ìôî (Ï∫îÎ∞îÏä§(fig)ÏôÄ ÎèÑÌôîÏßÄ(ax) Ï§ÄÎπÑÌïòÍ∏∞)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ -> seaborn ÏÇ¨Ïö©
sb.lineplot(x=[2, 3, 4, 5, 6], y=[10, 20, 30, 40, 50],
c="#2548b1", linestyle=':', linewidth=5,
marker="o", markersize=20, markerfacecolor="#00ff00",
markeredgecolor="#ff0000", markeredgewidth=3)

#3) Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
ax.set_title("Í∏∞Î≥∏Í∑∏Î¶¨Í∏∞", color="#000000", fontsize=22,
fontweight=1000, pad=20)
ax.set_xlabel("XÏ∂ï", color="#000000", fontsize=16, labelpad=10)
ax.set_ylabel("YÏ∂ï", color="#000000", fontsize=16, labelpad=10)

# 4) Ï∂úÎ†•
#plt.grid()
plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # Í∑∏ÎûòÌîÑ ÌôîÎ©¥ Ï∂úÎ†•
plt.close()
```
