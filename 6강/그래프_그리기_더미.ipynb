```python
### ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì •

#$ pip install --upgrade matplotlib seaborn (seaborn ì„¤ì¹˜)
from hossam import load_data
# ê¸€ê¼´ì„ ì‹œìŠ¤í…œì— ë“±ë¡
from matplotlib import font_manager as fm
# ìº”ë²„ìŠ¤(figure)ë¥¼ ìƒì„±, ê¸°ë³¸ ê·¸ëž˜í”½ í•¨ìˆ˜ ì œê³µ
from matplotlib import pyplot as plt
# ê³ ê¸‰ ê·¸ëž˜í”½ ê¸°ëŠ¥ ì œê³µ
import seaborn as sb
```

```python
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
# í•œê¸€ì„ ì§€ì›í•˜ëŠ” í°íŠ¸ íŒŒì¼
# í°íŠ¸ì˜ ê¸€ê¼´ì„ ì‹œìŠ¤í…œì— ë“±
font_prop = fm.FontProperties(fname=font_path) # í°íŠ¸ì˜ ì†ì„±ì„ ì½ì–´ì˜´
font_name = font_prop.get_name() # ì½ì–´ì˜¨ ì†ì„±ì—ì„œ í°íŠ¸ì˜ ì´ë¦„ë§Œ ì¶”ì¶œ
font_name
```

```python
my_dpi = 200
my_font_name = "Noto Sans KR" 
my_theme = "dark" # ì´ë¯¸ì§€ ì„ ëª…ë„ë¥¼ ê²°ì •í•˜ëŠ” 1ì¸ì¹˜
# ì‹œìŠ¤í…œì— ë“±ë¡ëœ ê¸€ê¼´ ì´ë¦„
# ê·¸ë¦¼ ìŠ¤íƒ€ì¼ ì§€ì •
sb.set_theme(style=my_theme) # seaborn ìŠ¤íƒ€ì¼ (í™”í’ ì„¤ì •í•˜ê¸°)
plt.rcParams['font.family'] = my_font_name # ê·¸ëž˜í”„ì— í•œê¸€ í°íŠ¸ ì ìš©
plt.rcParams['font.size'] = 16 # ê¸°ë³¸ í°íŠ¸ í¬ê¸°
plt.rcParams['axes.unicode_minus'] = False # ê·¸ëž˜í”„ì— ë§ˆì´ë„ˆìŠ¤ ê¹¨ì§ ë°©ì§€(í•œê¸€í™˜ê²½ì—ì„œ í•„ìˆ˜)ðŸ“˜
```

```python
# 1) ê·¸ëž˜í”„ ì´ˆê¸°í™” (ìº”ë°”ìŠ¤(fig)ì™€ ë„í™”ì§€(ax) ì¤€ë¹„í•˜ê¸°)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) ê·¸ëž˜í”„ ê·¸ë¦¬ê¸° -> seaborn ì‚¬ìš©
sb.lineplot(x=[],y=[])

#3) ê·¸ëž˜í”„ ê¾¸ë¯¸ê¸°
ax.set_title("ì œëª©ìž…ë‹ˆë‹¤", color="#000000", fontsize=22,fontweight=1000, pad=20)
ax.set_xlabel("Xì¶•", color="#000000", fontsize=16, labelpad=10)
ax.set_ylabel("Yì¶•", color="#000000", fontsize=16, labelpad=10)

# 4) ì¶œë ¥
#plt.grid()
plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # ê·¸ëž˜í”„ í™”ë©´ ì¶œë ¥
plt.close()
```
