### 패키지 참조

```python
from hossam import load_data
from matplotlib import font_manager as fm
from matplotlib import pyplot as plt
import seaborn as sb
import numpy as np
```

```python
my_dpi = 200
fpath = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(fpath)
fprop = fm.FontProperties(fname=fpath) # 폰트의 속성을 읽어옴
fname = fprop.get_name() # 읽어온 속성에서 폰트의 이름만 추출
plt.rcParams['font.family'] = fname
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data("penguins")
origin
```

#### #02. barplot

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.barplot(data=origin,
             x='species',
             y='body_mass_g',
             hue='sex',
             estimator=np.mean,
             errorbar=('ci',95),
             palette='Set2')

#3) 그래프 꾸미기
ax.grid(True)

# 4) 출력

plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # 그래프 화면 출력
plt.close()
```

#### #03. countplot (빈도 그래프)

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.countplot(data=origin,
             x='species',
             hue='sex',
             order=['Chinstrap','Gentoo','Adelie'],
             palette='Set2')

#3) 그래프 꾸미기
ax.grid(True)

# 4) 출력

plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # 그래프 화면 출력
plt.close()
```

#### #04. pointplot

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.pointplot(
                data=origin, # 사용할 데이터프레임
                x="species",
                y="flipper_length_mm",
                hue="sex",
                estimator=np.mean, # 요약 방식
                errorbar=("ci", 95), # 신뢰구간(기본 95%)
                linestyles="--", # "-", ".-", ":", "-."
                linewidth=1,
                markers="o", # 마커 종류: "o", "s", "D", "^", "v","*", ...
                palette="Dark2"
            )

#3) 그래프 꾸미기
ax.grid(True)

# 4) 출력

plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show() # 그래프 화면 출력
plt.close()
```
