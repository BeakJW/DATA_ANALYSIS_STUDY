```python
from hossam import load_data
from matplotlib import font_manager as fm
from matplotlib import pyplot as plt
import seaborn as sb
```

### 데이터 불러오기

```python
origin = load_data("traffic_acc")
origin
```

```python
df = origin.drop('월',axis=1).groupby('년도').sum()
df
```

```python
my_dpi = 200
fpath = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(fpath)
fprop = fm.FontProperties(fname=fpath) # 폰트의 속성을 읽어옴
fname = fprop.get_name() # 읽어온 속성에서 폰트의 이름만 추출
plt.rcParams['font.family'] = fname
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.lineplot(data=df, x=df.index,y='발생건수',
           color="#000000",linewidth=0.7, linestyle=':',
            marker="o", markersize=2,
            markerfacecolor="#303130",
            markeredgecolor="#ff0000", markeredgewidth=1)

#3) 그래프 꾸미기
ax.grid(True)
# 4) 출력

plt.tight_layout()
plt.show() # 그래프 화면 출력
plt.close()
```

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.lineplot(data=df, x=df.index, y='부상자수',label='부상자수')
sb.lineplot(data=df, x=df.index, y='발생건수',label='발생건수')

#3) 그래프 꾸미기
ax.grid(True)
# 4) 출력
plt.tight_layout()
plt.show() # 그래프 화면 출력
plt.close()
```

```python
df1 = df.reset_index()
df1
```

```python
df2 = df1.melt(id_vars='년도',value_vars=['발생건수','사망자수','부상자수'],var_name='구분')
df2
```

```python
# 1) 그래프 초기화 (캔바스(fig)와 도화지(ax) 준비하기)
width_px = 1000
height_px = 800
rows = 1
cols = 1
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#2) 그래프 그리기 -> seaborn 사용
sb.lineplot(data=df2, x='년도',y='value', hue='구분')

#3) 그래프 꾸미기
ax.grid(True)


plt.tight_layout()

plt.show() # 그래프 화면 출력
plt.close()
```
