### 배열 생성하기
#### 리스트를 통한 1차원(==1행으로 구성) 배열 만들기

```python
arr = np.array([1, 3, 5, 7, 9])
print(type(arr))
arr
```

### 배열의 크기
#### 'len()' 함수를 사용하여 배열의 크기를 확인 할 수 있다.

```python
arr = np.array([10, 20, 30, 40, 50])
size = len(arr)
```

```python
print("배열의 원소는 %d개 입니다" % size)
```

### 배열의 인덱싱과 슬라이싱
#### 기본 특성이 리스트와 동일하기 때문에 인덱싱과 슬라이싱이 가능하다.

```python
mydata = np.array([1, 2, 3, 4, 5])

print(mydata[0])
print(mydata[3])
```

```python
print(mydata[1:4])
```

### 반복문을 통한 활용
#### 리스트와 동일하게 사용할 수 있다.

```python
mydata = np.array([1, 2, 3, 4, 5])

for i, v in enumerate(mydata):
    print("%d번쨰의 원소 >> %d" % (i, v))
```
