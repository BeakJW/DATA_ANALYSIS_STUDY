## #01. Series 자료형

### 라이브러리 참조

```python
import numpy as np
from pandas import Series
```

### Series객체생성

```python
my_list = [10,30]
s1 = Series(my_list)
s1
```

```python
my_array = np.array([100,200])
s2 = Series(my_array)
s2
```

```python
my_dict = {"hello":123,"world":456,"nice":789}
s3 = Series(my_dict)
s3
```

### Series의정보확인

```python
my_series = Series({"hello":123,"world":456,"nice":789})
my_series.dtype
```

```python
my_series.info()
```

### Series객체의인덱싱과슬라이싱(1)

```python
my_series = Series([10,20,30,40,50])
print("==인덱싱==")
print(my_series[0])
print("==슬라이싱==")
print(my_series[1:3])
```

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})
print("==인덱싱==")
print(my_series["a"])
print("==슬라이싱==")
print(my_series["b":"d"])
```

### Series의값만추출하기

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})

print(type(my_series.values))
my_series.values
```

```python
mylist = list(my_series.values)
mylist
```

### Series의인덱스만추출하기

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})

print(type(my_series.index))
my_series.index
```

```python
mylist = list(my_series.index)
mylist
```
