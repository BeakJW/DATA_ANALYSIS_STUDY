## 5. 데이터 스케일링

### 라이브러리 참조

```python
from hossam import load_data
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, RobustScaler
```

```python
origin = load_data('nursing_grades')
origin
```

### #02.Min-MaxScaler(정규화)

```python
minmax_df = origin.copy()

x = minmax_df['필기점수']
xmin = x.min()
#print(x.min())
xmax = x.max()

minmax_df['필기점수_MinMax(1)'] = (x-xmin)/(xmax-xmin)
minmax_df.head()
```

```python
x = minmax_df[['필기점수']]
x.head()
```

```python
scaler = MinMaxScaler()
minmax_df['필기점수_MinMax(2)'] = scaler.fit_transform(x)
minmax_df.head()
```

### #03.Max-AbsScaler

```python
abs_df = origin.copy()
x = abs_df['필기점수']
abs_df['필기점수1'] = x/x.abs().max()
abs_df.head()
```

```python
scaler = MaxAbsScaler()
x = abs_df[['필기점수']]
abs_df['필기점수2'] = scaler.fit_transform(x)
abs_df.head()
```

### #04.StandardScaler(z-score,표준화)

```python
std_df = origin.copy()
x = origin['학부성적']
std_df['학부성적1'] = (x-x.mean())/x.std()
std_df.head()
```

```python
scaler = StandardScaler()
x = std_df[['학부성적']]
std_df['학부성적2'] = scaler.fit_transform(x)
std_df.head()
```

### 5. RobustScaler

```python
rb_df = origin.copy()

x = rb_df['병원경력']
m = x.median()
iqr = x.quantile(0.75) - x.quantile(0.25)

rb_df['병원경력2'] = (x-m) / iqr
rb_df.head()
```

```python
scaler = RobustScaler()
x = rb_df[['병원경력']]
rb_df['병원경력2'] = scaler.fit_transform(x)
rb_df.head()
```
