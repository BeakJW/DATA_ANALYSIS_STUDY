```python
df4 = origin.copy()
df4['í•©ê³„'] = df4.iloc[:,2:6].sum(axis=1)
df4['í‰ê· '] = df4.iloc[:,2:6].mean(axis=1)
df4
```

```python
df4['ê²°ê³¼'] = np.where(df4['í‰ê· ']>=80,'í•©ê²©','ë¶ˆí•©ê²©')
df4
```

```python
df4['ê²°ê³¼2'] = 'ë¶ˆí•©ê²©'
df4.loc[df4['í‰ê· ']>=80,'ê²°ê³¼2'] = 'í•©ê²©'
df4
```

```python
conditions = [
    (df4['í‰ê· ']>=90),
    (df4['í‰ê· ']>=80),
    (df4['í‰ê· ']>=70),
]

value = ['A','B','C']

df4['í•™ì '] = np.select(conditions,value,default='F')
df4
```

```python
def getScore(x):
    if x >=90:
        score = 'A'
    elif x>=80:
        score = 'B'
    elif x>=70:
        score = 'C'
    else:
        score = 'F'
    return score
```

```python
df4['í•™ì 2'] = df4['í‰ê· '].apply(getScore)
df4
```

```python
df4['í•™ì 3'] = df4['í‰ê· '].apply(
    lambda x: {10:'A',9:'A',8:'B',7:'C'}.get(x//10, 'F')
)
df4
```
