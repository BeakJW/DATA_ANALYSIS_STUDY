### 조건식에따른원소추출

```python
mypoint = np.array([82, 77, 91, 88])
ge80 = mypoint[mypoint >= 80]
ge80
```

```python
ge80lee90 = mypoint[np.logical_and(mypoint >= 80, mypoint <= 90)]
ge80lee90
```

```python
gt_or_lt = mypoint[np.logical_or(mypoint < 80, mypoint > 90)]
gt_or_lt
```

### NumPy의조건연산

```python
exam = np.array([60, 83, 75, 92, 86])

result = np.where(exam >= 80, "합격", "불합격")
result
```

### 여러 조건을 한번에 처리하기

```python
exam = np.array([60, 83, 75, 92, 86])

conditions = [
    (exam >= 90),  # if
    (exam >= 80),  # elif
    (exam >= 70),  # elif
]

value = ["A", "B", "C"]

result = np.select(conditions, value, default="F")
result
```
