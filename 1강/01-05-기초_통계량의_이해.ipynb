### ë°ì´í„°ì˜ìœ„ì¹˜Â·ë¶„í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(2)

### ë°ì´í„°ì˜ìœ„ì¹˜Â·ë¶„í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(4)
#### â˜ ì¤‘ì•™ê°’ (ì§ì ‘ êµ¬í•˜ê¸°)

```python
size = len(myarr) # ì „ì²´ ë°ì´í„° ìˆ˜
sorted_arr = np.sort(myarr) # ìˆœì„œëŒ€ë¡œ ì •ë ¬
print("ì •ë ¬:",sorted_arr)

#ì§ìˆ˜ì¸ ê²½ìš° ê°€ìš´ë° ë‘ ìˆ˜ì˜ í‰ê· 
if size%2==0:
    p = (size//2) - 1 # ì¸ë±ìŠ¤ê°€ 0ë¶€í„° ì‹œì‘í•˜ë¯€ë¡œ -1
    q = p+1 #pë²ˆì§¸ ë‹¤ìŒ ìš”ì†Œì˜ ì¸ë±ìŠ¤
    median = (sorted_arr[p]+sorted_arr[q])/2
else:
    p = size//2
    median = sorted_arr[p]

print("ì¤‘ì•™ê°’:",median)
```

### ë°ì´í„°ì˜ìœ„ì¹˜Â·ë¶„í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(5)
#### â˜ ì¤‘ì•™ê°’ (NumPyAPIì‚¬ìš©)

```python
print("ì¤‘ì•™ê°’:",np.median(myarr))
```

### ë°ì´í„°ì˜ìœ„ì¹˜Â·ë¶„í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(9)
#### 1ì‚¬ë¶„ìœ„ ìˆ˜

```python
x = np.sort(myarr)
n = len(x)
q1_pos = (n-1)*0.25
i = int(np.floor(q1_pos))
f = q1_pos - i

#ë³´ê°„
Q1 = x[i] + f *(x[i+1]-x[i])
print("ì§ì ‘ ê³„ì‚°í•œ 1ì‚¬ë¶„ìœ„ìˆ˜ (Q1):",Q1)

#numpy ë‚´ì¥ í•¨ìˆ˜ì™€ ë¹„êµ
print("Numpyë¡œ ê³„ì‚°í•œ 1ì‚¬ë¶„ìœ„ìˆ˜:",np.quantile(myarr,0.25))
```

### ë°ì´í„°ì˜ìœ„ì¹˜Â·ë¶„í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(12)
#### 3ì‚¬ë¶„ìœ„ ìˆ˜

```python
x = np.sort(myarr)
n = len(x)
q3_pos = (n-1)*0.75
i = int(np.floor(q3_pos))
f = q3_pos - i

#ë³´ê°„
Q3 = x[i]+f*(x[i+1]-x[i])
print("ì§ì ‘ ê³„ì‚°í•œ 1ì‚¬ë¶„ìœ„ ìˆ˜(Q1)",Q3)

#numpy ë‚´ì¥ í•¨ìˆ˜ì™€ ë¹„êµ
print("Numpyë¡œ ê³„ì‚¬ë‚œ 1ì‚¬ë¶„ìœ„ìˆ˜:",np.quantile(myarr,0.75))
```

### ë°ì´í„°ì˜ë²”ìœ„Â·ì‚°í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(1)

### ë°ì´í„°ì˜ë²”ìœ„Â·ì‚°í¬ë¥¼ë‚˜íƒ€ë‚´ëŠ”ê¸°ì´ˆí†µê³„ëŸ‰(2)
