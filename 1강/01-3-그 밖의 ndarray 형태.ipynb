### 2차원 배열

```python
import numpy as np

myarray = np.array([[98, 12, 23, 54], [32, 53, 21, 76], [12, 67, 87, 43]])

print(myarray)

print("차원크기", myarray.ndim)
print("차원별 원소수", myarray.shape)
print("원소의 데이터 타입:", myarray.dtype)
```

### boolean타입으로인덱싱

```python
bool_array = np.array(
    [[True, False, True, False], [True, True, True, False], [True, True, True, False]]
)
```

```python
result1 = myarray = myarray[bool_array]
result1
```

### 비교식에의한검색

```python
result2 = myarray[myarray >= 80]
result2
```

```python
result3 = myarray[np.logical_and(myarray >= 80, myarray <= 90)]
result3
```

```python
result4 = myarray[np.logical_or(myarray < 80, myarray > 90)]
result4
```
