### 라이브러리 참조

```python
import numpy as np
```

### 샘플 데이터 준비

```python
point = np.array([100,91,89,86,84,79,78,77,74,71,69,66,65,60,58,57,55])
point
```

### Numpy의histogram:도수분포도(1)
#### ☐ 5개의 구간으로 나누기

```python
hist,bins = np.histogram(point,5)
print("구간별 데이터 수",hist)
print("구간 경계:",bins)
```

#### 도수 분포표 구성

```python
s=len(hist)

for i in range(0,5):
    if i +1 <s:
        tpl = "%d이상 %d미만: %d개"
    else:
        tpl = "%d이상 %d이하: %d개"
    print(tpl%(bins[i],bins[i+1],hist[i]))
```

#### 데이터 구간을 직접 정의하기

```python
hist,bins = np.histogram(point,[50,60,70,80,90,100])
print("구간별 데이터 수",hist)
print("구간 경계:",bins)
```

### 도수의 종류

#### 절대 도수

```python
#np.histogram에서 리턴하는 hist값이 절대도수
# 상대도수
relative_freq = hist /np.sum(hist)
relative_freq
```

#### 누적 도수

```python
cum_freq = []
previous = 0

for i in hist:
    previous += i 
    cum_freq.append(previous)
previous
```
